```python
clean_data=data.dropna(axis='columns')
```

```python
len(data.columns),len(clean_data.columns)
```

```python
data_plot=(clean_data.loc[2018]-clean_data.loc[1998])/clean_data.loc[1998]
```

```python
data_plot.describe()
```

```python
data_plot.plot.hist(bins=30)
```

```python
(data_plot<0).value_counts()
```

```python
(data_plot<0).value_counts().plot.pie(colors=['r','b'],labels=['>=0','<0'],title='TITLE',ylabel='label')
```

```python
(data_plot<0).value_counts().plot.pie(colors=['r','b'],labels=['>=0','<0'],title='countries with decrease in c02 per capita',ylabel='20 years co2 per capita development')
```
